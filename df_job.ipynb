```python
import pandas as pd

from transform import remove_html, maintain_alpha, remove_single, remove_morespace, repair_salary
```

```python
df_job = pd.read_csv('data/df_job.csv', index_col=['JobID'])
df_job.drop(columns=['Unnamed: 0', 'PostedDate', 'PublishDate'], inplace=True)

df_job[df_job.select_dtypes(object).columns] = df_job.select_dtypes(object).fillna(0)
df_job[df_job.select_dtypes(object).columns] = df_job.select_dtypes(object).astype(str)
```

```python
df_job.head(2)
```

```python
df_job.Description = df_job.Description.apply(
    lambda x: remove_morespace(
                remove_single(
                    maintain_alpha(
                        remove_html(
                            str.lower(x)
                        )
                    )
                )
    ).strip()
)
```

```python
df_job = df_job[df_job.JobStatus != 'Not Publish']
```

```python
df_job.head(2)
```

```python
df_job.JobTitle = df_job.JobTitle.map(str.lower).map(maintain_alpha).map(remove_morespace).map(str.strip)
```

```python
df_job.head(2)
```

```python
df_job.Requirement = df_job.Requirement.map(remove_html).map(maintain_alpha).map(remove_single).map(remove_morespace).map(str.lower).map(str.strip)
```

```python
df_job.head(1)
```

```python
df_job[['SalaryMax', 'SalaryMin']] = df_job[['SalaryMax', 'SalaryMin']].astype(int)
```

```python
df_job.drop(columns=['UserId'], inplace=True)
```

```python
df_job.drop(columns=['ClosedDate'], inplace=True)
```

```python
df_job.DriverLicenseType = df_job.DriverLicenseType.map(str.lower)
```

```python
df_job.Gender = df_job.Gender.fillna(0).astype(int)
```

```python
df_job.MaritalStatus = df_job.MaritalStatus.fillna(0).astype(int)
```
