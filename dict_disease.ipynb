```python
DRIVER_PATH = 'C:/users/febri/projects/eris_job/driver/chromedriver.exe'
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
driver.get('https://www.guesehat.com/nama-penyakit-a-z')
```

```python
raw_ill = driver.find_elements_by_xpath('/html/body/div[2]/div[2]/section[2]/div/div/ul')[0].text
```

```python
new_ill = [i.replace(')', '').strip() for i in '('.join(raw_ill.split('\n')).split('(')]
```

```python
the_ill = []
for ill in new_ill:
    ills = ill.split('/')
    if len(ills) > 1:
        ills = [i.strip() for i in ills]
        ills = '/'.join(ills)
    else:
        ills = ills[0]
    the_ill.append(ills)
```

```python
the_ill = [i.lower() for i in the_ill]
```

```python
new_ill = []
for ill in the_ill:
    ills = ill.split('/')
    if len(ills) > 1:
        for il in ills:
            new_ill.append(il)
    else:
        new_ill.append(ill)
```

```python
the_ill = []

for ill in new_ill:
    ill = ill.replace('\'', '')
    ill = ill.replace('-', '')
    ill = ill.replace(',', '')

    the_ill.append(ill)
```

```python
pd.DataFrame(the_ill, columns=['disease']).to_csv('dict_disease_original.csv', index=False)
```

```python
df_applicant = pd.read_csv('data/df_applicant.csv')

df_applicant.drop(columns=['Unnamed: 0'], inplace=True)
df_applicant.set_index(['ApplicantID'], inplace=True)

df_applicant = df_applicant.fillna(0)
df_applicant.ExpectedSalary = df_applicant.ExpectedSalary.apply(lambda x: 0 if x < 1_000_000 else x)

df_app_obj = df_applicant.select_dtypes(object).astype(str).applymap(str.lower)
df_app_int = df_applicant.select_dtypes(int)
df_app_flt = df_applicant.select_dtypes(float).astype(int)

df_applicant = pd.concat([df_app_obj, df_app_int, df_app_flt], axis=1)
df_applicant.DiseaseHistory = df_applicant.DiseaseHistory.apply(lambda x: str(0) if not (all(chr.isalpha() or chr.isspace() for chr in x) and len(x) != 1) else x.strip())
```

```python
# pembuatan kamus DiseaseHistory
setwords_DiseaseHistory = []

for penyakits in df_applicant.DiseaseHistory.unique():
    penyakits = penyakits.split(' ')
    if len(penyakits) > 1:
        for penyakit in penyakits:
            setwords_DiseaseHistory.append(penyakit)
    else:
        setwords_DiseaseHistory.append(penyakits[0])
else:
    setwords_DiseaseHistory = list(set([i for i in setwords_DiseaseHistory if len(i) != 1]))
```
